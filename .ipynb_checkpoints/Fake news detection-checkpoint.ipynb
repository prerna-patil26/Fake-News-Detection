#### **Text Preprocessing & Cleaning**

```python
def wordopt(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub("\\W"," ",text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text

df['text'] = df['text'].apply(wordopt)
```

→ Lowercasing: Converted all text to lowercase to maintain consistency.             
→ URL Removal: Removed links (http:// or https://) as they do not add value to text analysis.                 
→ Mentions & Hashtags Removal: Removed @user and #topic as they are not useful for classification.                     
→ Punctuation & Special Characters Removal: Kept only alphanumeric characters and spaces for cleaner text.          
→ Tokenization: Split text into individual words for further processing.            
→ Stopwords Removal: Removed common words like "the," "is," "and" that do not contribute to meaning.               
→ Lemmatization: Converted words to their root forms (e.g., "running" → "run") to normalize text.

### 3️⃣ Train-Test Split & TF-IDF ###

#### Converting Text to Numbers (TF-IDF)

##### 1️⃣ TF (Term Frequency) – Counts how often a word appears in a document.
##### - Example:
        In "Apple is a fruit. Apple is sweet." → TF("Apple") = 2

##### 2️⃣ IDF (Inverse Document Frequency) – Measures how rare a word is across documents.
##### - Example:
            Common word ("is") → Low IDF
            Rare word ("Dragonfruit") → High IDF
            
🔹 TF-IDF helps models focus on important words while ignoring common ones!

##### ➡️ Why did we use TF-IDF?

###### -Removes common words bias: It reduces the weight of frequently used words like "the", "is", "and", which do not provide meaningful information.
###### -Highlights important words: It gives more weight to words that are unique to a document and might help differentiate between fake and real news.
###### -Improves model performance: Since the dataset is converted into numerical form, ML algorithms work better and provide more accurate results.
###### -Better than simple word count: Unlike basic word frequency methods, TF-IDF balances common and rare words smartly, improving text analysis.



##### ➡️How is it useful for our model?
###### -Using TF-IDF, our model can understand which words are more important for classification. This helps it identify patterns in fake and real news more effectively, making predictions more reliable.

#### **Function for create confusion matrix , precision reacall , Roc_Curve**

```python
# Your custom functions
def get_model_scores(model, x_test):
    if hasattr(model, "predict_proba"):  # If model supports predict_proba()
        return model.predict_proba(x_test)[:, 1]
    elif hasattr(model, "decision_function"):  # If model has decision_function()
        return model.decision_function(x_test)
    else:
        raise AttributeError(f"{model.__class__.__name__} does not support probability estimation.")
```

```python
def plot_roc_curve(ax, model, x_test, y_test):
    y_scores = get_model_scores(model, x_test)
    fpr, tpr, _ = roc_curve(y_test, y_scores)
    roc_auc = auc(fpr, tpr)
    ax.plot(fpr, tpr, color='blue', lw=2, label=f'AUC = {roc_auc:.3f}')
    ax.plot([0, 1], [0, 1], color='gray', linestyle='--')
    ax.set_xlabel("False Positive Rate")
    ax.set_ylabel("True Positive Rate")
    ax.set_title("ROC Curve")
    ax.legend(loc="lower right")
    return roc_auc
```

```python
def plot_precision_recall_curve(ax, model, x_test, y_test):
    y_scores = get_model_scores(model, x_test)  # ✅ Automatically selects correct method
    precision, recall, _ = precision_recall_curve(y_test, y_scores)
    ax.plot(recall, precision, color='red', lw=2)
    ax.set_xlabel("Recall")
    ax.set_ylabel("Precision")
    ax.set_title("Precision-Recall Curve")
```

```python
# Function to plot Confusion Matrix
def plot_confusion_matrix(ax, y_test, y_pred, title):
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", ax=ax)
    ax.set_xlabel("Predicted")
    ax.set_ylabel("Actual")
    ax.set_title(title)
```

### **Ensemble Model (Voting Classifier)**

```python
from sklearn.ensemble import VotingClassifier
```

```python
# Create an ensemble model using Voting Classifier with hard voting
ensemble = VotingClassifier(
    estimators=[
        ('rf', rf),  # Random Forest
        ('lr', lr),  # Logistic Regression
        ('nb', nb),  # Naive Bayes
        ('dt', dt)   # Decision Tree
    ],
    voting='hard'  # Hard voting for majority decision
)

# Fit the model on training data
ensemble.fit(xv_train, y_train)

# Predictions
y_pred_ensemble = ensemble.predict(xv_test)

# Print results
print("\n🔹 Ensemble Model Results:")
print("Train Accuracy:", accuracy_score(y_train, ensemble.predict(xv_train)))
print("Test Accuracy:", accuracy_score(y_test, y_pred_ensemble))
print("Classification Report:\n", classification_report(y_test, y_pred_ensemble))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_ensemble))

# Plot Graphs
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
plot_confusion_matrix(axes[0], y_test, y_pred_ensemble, title="Ensemble Model Confusion Matrix")

# Manually calculate ROC curve for hard voting
from sklearn.metrics import roc_curve, auc

# Get predictions from each model
y_pred_rf = rf.predict(xv_test)
y_pred_lr = lr.predict(xv_test)
y_pred_nb = nb.predict(xv_test)
y_pred_dt = dt.predict(xv_test)

# Combine predictions using majority voting
y_pred_combined = np.array([y_pred_rf, y_pred_lr, y_pred_nb, y_pred_dt])
y_pred_majority = np.apply_along_axis(lambda x: np.bincount(x).argmax(), axis=0, arr=y_pred_combined)

# Calculate ROC curve and AUC
fpr, tpr, _ = roc_curve(y_test, y_pred_majority)
roc_auc = auc(fpr, tpr)

# Plot ROC curve
axes[1].plot(fpr, tpr, color='blue', lw=2, label=f'AUC = {roc_auc:.3f}')
axes[1].plot([0, 1], [0, 1], color='gray', linestyle='--')
axes[1].set_xlabel("False Positive Rate")
axes[1].set_ylabel("True Positive Rate")
axes[1].set_title("ROC Curve")
axes[1].legend(loc="lower right")

# Plot Precision-Recall Curve
precision, recall, _ = precision_recall_curve(y_test, y_pred_majority)
axes[2].plot(recall, precision, color='red', lw=2)
axes[2].set_xlabel("Recall")
axes[2].set_ylabel("Precision")
axes[2].set_title("Precision-Recall Curve")

print(f"Ensemble Model AUC Score: {roc_auc:.3f}")
plt.show()
```

### 5️⃣ Model Training & Evaluation ###

```python
from sklearn.metrics import accuracy_score, roc_auc_score
import pandas as pd

# Convert y_test to integer (if not already)
y_test = y_test.astype(int)

# Dictionary to store model performance
model_performance = {}

# Your selected models
models = {
    'Logistic Regression': lr,  # Logistic Regression
    'Passive Aggressive': pa,   # Passive Aggressive
    'Random Forest': rf,        # Random Forest
    'Naive Bayes': nb,          # Naive Bayes
    'Decision Tree': dt,        # Decision Tree
    'XGBoost': xgb              # XGBoost
}

# Evaluate each model
for name, model in models.items():
    y_pred = model.predict(xv_test)  # Predictions
    accuracy = accuracy_score(y_test, y_pred)  # Accuracy

    # Check if the model supports predict_proba for AUC calculation
    if hasattr(model, 'predict_proba'):
        auc = roc_auc_score(y_test, model.predict_proba(xv_test)[:, 1])
    else:
        auc = 0  # If predict_proba is not available, set AUC to 0

    # Store performance metrics
    model_performance[name] = {'Accuracy': accuracy, 'AUC': auc}

# Convert performance data to a DataFrame
performance_df = pd.DataFrame(model_performance).T
print("\n🔹 Model Performance:\n", performance_df)

# Identify the best model based on the average of Accuracy and AUC
best_model = performance_df[['Accuracy', 'AUC']].mean(axis=1).idxmax()
print(f"\n Best Model: {best_model}")

# Extract model names and their accuracy scores
model_names = list(model_performance.keys())
accuracy_scores = [model_performance[model]['Accuracy'] * 100 for model in model_names]  # Convert to percentage
```

## ➡️**Using LIME (Explanation) Predicion:**

```python
from sklearn.pipeline import make_pipeline

# Create a pipeline with TF-IDF and the ensemble model
pipeline = make_pipeline(vectorization, ensemble)  # Use ensemble model here
explainer = lime.lime_text.LimeTextExplainer(class_names=['Real', 'Fake'])  # Class names updated

def predict_proba_approximation(model, X):
    """
    Approximate probabilities for hard voting ensemble.
    """
    # Get predictions from each model in the ensemble
    predictions = np.array([m.predict(X) for m in model.estimators_])

    # Calculate the fraction of models predicting each class
    proba = np.array([np.mean(predictions == i, axis=0) for i in range(2)]).T
    return proba

def explain_news(news_text):
    # Transform the input text using the TF-IDF vectorizer
    news_text_tfidf = vectorization.transform([news_text])

    # Get the model's prediction
    prediction = ensemble.predict(news_text_tfidf)[0]

    # Approximate probabilities
    prediction_prob = predict_proba_approximation(ensemble, news_text_tfidf)[0]

    # Generate LIME explanation
    exp = explainer.explain_instance(
        news_text,
        lambda x: predict_proba_approximation(ensemble, vectorization.transform(x)),
        num_features=10
    )

    # Show explanation
    print("\n🔹 **Model Prediction:**", "Real News" if prediction == 0 else "Fake News")  # Updated labels
    print(f"🔸 Probability: Real = {prediction_prob[0]:.2f}, Fake = {prediction_prob[1]:.2f}")  # Updated labels
    exp.show_in_notebook(text=True)  # Shows explanation in Jupyter Notebook

# Take user input and explain the prediction
user_input = input("Enter a news article: ")
explain_news(user_input)
```

```python
def output_label(n) :
  if n==0:
    return "It is a fake news"
  elif n==1:
    return "It is a real news"
  else:
    return "Invalid value"
```

```python
def manual_testing(news):
  testing_news = {"text":[news]}
  new_def_test = pd.DataFrame(testing_news)
  new_def_test["text"] = new_def_test["text"].apply(wordopt)
  new_x_test = new_def_test["text"]
  new_xv_test = vectorization.transform(new_x_test)
  pred_LR = lr.predict(new_xv_test)
  pred_DT = dt.predict(new_xv_test)
  pred_GB = gb.predict(new_xv_test)
  pred_PA = pa.predict(new_xv_test)
  pred_NB = nb.predict(new_xv_test)
  pred_RF = rf.predict(new_xv_test)
  pred_XGB = xgb.predict(new_xv_test)
  print("\n🔹 Logistic Regression Prediction:", output_label(pred_LR[0]))
  print("🔹 Decision Tree Prediction:", output_label(pred_DT[0]))
  print("🔹 Gradient Boosting Prediction:", output_label(pred_GB[0]))
  print("🔹 Passive Aggressive Prediction:", output_label(pred_PA[0]))
  print("🔹 Naive Bayes Prediction:", output_label(pred_NB[0]))
  print("🔹 Random Forest Prediction:", output_label(pred_RF[0]))
  print("🔹 XGBoost Prediction:", output_label(pred_XGB[0]))
  return
```
